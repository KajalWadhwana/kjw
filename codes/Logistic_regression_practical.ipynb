```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```
dataset=pd.read_csv("/content/drive/MyDrive/logistic regression cancer/breast-cancer.csv", header = None)
x=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```
x[6].value_counts()
x[6].replace("?",x[6].mode().iloc[0], inplace= True)
x[6].value_counts()
```

```
x=x.drop([0],axis=1)
x
```

```
x.shape
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=1)
```

```
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()
classifier.fit(x_train,y_train)
```

```
y_pred=classifier.predict(x_test)
```

```
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

# K-fold cross validation

**its a technique where your model will by default generate 10 accuracies and then you can see the value of mean of all the accuracies and also the standard deviation**

```
from sklearn.model_selection import cross_val_score
accuracies = cross_val_score(classifier, x_train, y_train, cv = 10)
print("accuracy: {:.2f} %" .format(accuracies.mean()*100))
print("standard deviation: {:.2f} %" .format(accuracies.std()*100))
```
